# Extend density profiles when there is no pedestal (L-mode-like plasmas)

```python
# Working directory (must contain the original data files)
working_dir = 'path to the working directory where the input equilibrium file can be found'
```

```python
profile = 'ne.dat' # either ne.dat or Te.dat
```

```python
# Import statements
from pylab import *
import math
from scipy.optimize import curve_fit

%matplotlib inline
```

```python
# Load density data
data = loadtxt(working_dir+'/'+ profile, skiprows=1)
rho = data[:,0] # rho poloidal
prf = data[:,1] # profile (either ne or Te)
```

```python
figure(0, figsize=(12,4))
subplot(121)
plot(rho, prf, 'x');
xlabel(r'$\rho$');
subplot(122)
plot(prf);
xlabel('n. points');
savefig('Density.png')
```

```python
IndexStart = 40
drho = 0.01
nmbrAdditionalPoints = 200
scale = 1.0
```

```python
# Extrapolation function  
def func(scale,DeltaRho,amplitude,i0,imax,):
    new_rho = empty([imax])
    new_prf = empty([imax])
    for i in range(0,imax):
        if i >= i0:
            new_rho[i] = new_rho[i-1] + drho
            new_prf[i] = scale*amplitude * exp(-(new_rho[i] - rho[i0])/DeltaRho)
        else:
            new_rho[i] = rho[i]
            new_prf[i] = scale*prf[i]
    return new_rho, new_prf
```

```python
amplitude = prf[IndexStart]
DeltaRho = -amplitude*(rho[IndexStart] - rho[IndexStart-1]) / (prf[IndexStart] - prf[IndexStart-1])
new_rho, new_prf = func(scale,DeltaRho, amplitude, IndexStart, IndexStart+nmbrAdditionalPoints)

print(new_rho.shape, new_prf.shape)
```

```python
figure(1, figsize=(12,4))
# Density
subplot(121)
plot(rho, prf, 'x');
plot(new_rho, new_prf);
xlabel(r'$\rho$')
legend(('original data', 'extended data'));
```

```python
# Output files
filename = working_dir + '/' + profile + '.extended'

# Open output files
new_file = open(filename, 'ab')

# Clear the content of both files when not already empty
new_file.seek(0); new_file.truncate()

# Export the datane_55394_indThomson_42
data = array([new_rho, new_prf])
savetxt(new_file, array([new_rho.size], dtype=int), fmt='%i')
savetxt(new_file, data.T)

# Close files
new_file.close()
```
