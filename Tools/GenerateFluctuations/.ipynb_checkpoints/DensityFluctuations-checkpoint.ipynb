# A generator of a random scalar field with given correlation

The problem is generating a possibly large number of samples of a scalar random field of which we know the spectrum of two-point correlation function only. The field is meant to represent turbulent fluctuations of the particle density in a magnetized plasma. As such, we can consider the scalar field to be defined on a two-dimensional domain, as turbulence in magnetized plasmas is highly anisotropic with a very small wave number in the direction of the equilibrium magnetic field. 

## General set-up

We consider a two-dimensional box $\Omega = [x_1,x_2] \times [z_1, z_2]$ on the $(x,z)$-plane, together with a probability space $(A, \mathcal{F}, \mu)$ with probability measure $d\mu(\alpha)$ where $\alpha \in A$. A real-valued random field over $\Omega$ is a measurable function

\begin{equation*}
    \delta n : \Omega \times A \to \mathbb{R}.
\end{equation*}

We choose to represent such a field in terms of normalized coordinates

\begin{align*}
    \xi &= \frac{x - x_1}{x_2 - x_1} = \frac{x - x_1}{a_x}, \\
    \zeta &= \frac{z - z_1}{z_2 - z_1} = \frac{z - z_1}{a_z},
\end{align*}

and to make use of a truncated Fourier-like series representation

\begin{equation*}
    \delta n(x,z) = \sum_{k = -M_x}^{M_x} \sum_{\ell = -M_z}^{M_z}
    \widehat{\delta n}_{k,\ell}(x,z,\alpha) e^{2\pi i (k \xi + \ell \zeta)},
\end{equation*}

where the dependence on $\alpha \in A$ is implied in $\delta n(x,z)$. Here, $M_x, M_z$ are positive integers representing the maximum frequency present in the spectrum of the field. One can also notice that the spectral components $\widehat{\delta n}_{k,\ell}(x,z,\alpha)$, in addition to their natural dependence on $\alpha \in A$, are allowed to depend on the point location $(x,z) \in \Omega$, thus this is not strictly speaking a Fourier series. The choice of the probability space $A$ is crucial to the statistics of the field. The only criteria for the choice of $A$ and the amplitudes $\widehat{\delta n}_{k,\ell}(x,z,\alpha)$ are _(i)_ the fact that the field 
$\delta n(x,z)$ is real valued, hence,

\begin{equation*}
    \overline{\widehat{\delta n}_{k,\ell}(x,z,\alpha)} =
    \widehat{\delta n}_{-k,-\ell}(x,z,\alpha),
\end{equation*}

with over-lines denoting complex conjugation; _(ii)_ the fact that the fluctuations should not change the background and thus average out to zero; _(iii)_ the Wigner funciton of the two-point correlation, which is known from either turbulence simulation or experimental observations.

By definition, the two-point correlation function of a random field $\delta n(x,y)$ is a deterministic function $C : \Omega \times \Omega \to \mathbb{R}$ defined by

\begin{equation*}
    C(x,z,x',z') = \mathbb{E} \big( \delta n(x,z) \delta n(x',z')\big),
\end{equation*}

where $\mathbb{E}$ is the expectation value operator on $(A, \mathcal{F}, \mu)$. For the proposed model we have

\begin{equation*}
    C(x,z,x',z') = \sum_{k = -M_x}^{M_x} \sum_{\ell = -M_z}^{M_z}
    \sum_{k' = -M_x}^{M_x} \sum_{\ell' = -M_z}^{M_z}
    \mathbb{E}\big( \widehat{\delta n}_{k,\ell}(x,z, \cdot) 
        \widehat{\delta n}_{k',\ell'}(x',z',\cdot) \big)
    e^{2\pi i (k \xi + \ell \zeta)} e^{2\pi i (k'\xi' + \ell \zeta')},
\end{equation*}

where $\xi', \zeta'$ are the normalized coordinates corresponding to the point $(x',z')$. 

Upon assuming that the fluctuations and the corresponding correlation function are extended to the whole space by periodicity, the Wigner function is defined by

\begin{equation*}
    \Gamma(k_0,x,z,N_x,N_z) = \int e^{-ik_0 (N_x s_x + N_z s_z)}
    C(x+\tfrac{1}{2}s_x, z+\tfrac{1}{2}s_z ,x-\tfrac{1}{2}s_x,z-\tfrac{1}{2}s_z)ds_xds_z,
\end{equation*}

where $k_0 = \omega /c$ is the wave vector in free space of a monochromatic electromagnetic waves of frequency $\omega$ and $c$ is the speed of light. The variable $(N_x, N_z)$ is the wave refractive index. The integral is not absolutely convergent and must be treated in sense of distributions.

### Imposing the condition of real-valued fields

In one dimension, the fact that the field is real valued means that we only need to specify the amplitude of, e.g., non-negative frequencies. In two or more dimensions, there is a certain degree of freedom in choosing the independent harmonics. One can isolate the "zero frequencies" by rewriting the Fourier series in the form 

\begin{align*}
    \delta n(x,z) &= \widehat{\delta n}_{0,0}(x,z,\alpha) \\
    &\quad +\sum_{k = 1}^{M_x} \Big[ \widehat{\delta n}_{k,0} (x,z,\alpha) e^{2\pi i k\xi} +
        \widehat{\delta n}_{-k,0} (x,z,\alpha) e^{-2\pi i k\xi} \Big] \\
    &\quad +\sum_{\ell=1}^{M_z} \Big[ \widehat{\delta n}_{0,\ell} (x,z,\alpha) 
        e^{2\pi i \ell\zeta} +
        \widehat{\delta n}_{0,-\ell} (x,z,\alpha) e^{-2\pi i \ell\zeta} \Big] \\
    &\quad +\sum_{k=1}^{M_x} \sum_{\ell=1}^{M_z} \Big[ \widehat{\delta n}_{k,\ell} (x,z,\alpha)
        e^{2\pi i (k\xi + \ell\zeta)} + \widehat{\delta n}_{-k, -\ell} (x,z,\alpha) 
        e^{-2\pi i (k\xi + \ell\zeta)} \Big] \\
    &\quad +\sum_{k=1}^{M_x} \sum_{\ell=1}^{M_z} \Big[ \widehat{\delta n}_{-k,\ell} (x,z,\alpha)
        e^{2\pi i (k\xi + \ell\zeta)} + \widehat{\delta n}_{k, -\ell} (x,z,\alpha)
        e^{-2\pi i (k\xi + \ell\zeta)} \Big].
\end{align*}

The reality condition determines the second term in each sum given the first. We can single out the $(0,0)$ harmonic which does not have a corresponding "negative harmonic" and identify the set of the remaining statistically independent harmonics $\mathcal{M}$ as the union of four index sets, namely,

\begin{equation*}
    \mathcal{M} = \mathcal{M}_1 \cup \mathcal{M}_2 \cup \mathcal{M}_3 \cup \mathcal{M}_4,
\end{equation*}

where

\begin{align*}
    \mathcal{M}_1 &= \{ (k,0)     \;|\; 1 \leq k \leq M_x \}, \\
    \mathcal{M}_2 &= \{ (0,\ell)  \;|\; 1 \leq \ell \leq M_z \}, \\
    \mathcal{M}_3 &= \{ (k,\ell)  \;|\; 1 \leq k \leq M_x, 1 \leq \ell \leq M_z \}, \\ 
    \mathcal{M}_4 &= \{ (-k,\ell) \;|\; 1 \leq k \leq M_x, 1 \leq \ell \leq M_z \}.
\end{align*}

The field $\delta n(x,z)$ of random fluctuations is then represented as

\begin{equation*}
    \delta n (x,z) = \widehat{\delta n}_{0,0} (x,z,\alpha) +
    \sum_{(k,\ell) \in \mathcal{M}} \Big[ \widehat{\delta n}_{k,\ell} (x,z, \alpha)
    e^{2\pi i (k\xi + \ell \zeta)} + \text{c.c} \Big],
\end{equation*}

where "$\text{c.c.}$" stands for complex conjugate. This expression remove redundant degree of freedom and make it explicit that the field is real-valued.

For real-valued fields, we can write the correlation in terms of the statistically independent harmonics only, namely, 

\begin{align*}
    C(x,z,x',z') = &\mathbb{E}\big(\widehat{\delta n}_{0,0} (x,z,\cdot) 
        \widehat{\delta n}_{0,0}(x,z,\cdot)\big) \\
    &+ \sum_{(k,\ell)\in\mathcal{M}} \Big[ 
    \mathbb{E}\big( \widehat{\delta n}_{0,0} (x,z,\cdot) 
        \widehat{\delta n}_{k,\ell}(x',z',\cdot)\big)
    e^{2\pi i (k\xi + \ell\zeta)} \\
    &\qquad\qquad +\mathbb{E}\big( \widehat{\delta n}_{0,0} (x,z,\cdot) 
        \overline{\widehat{\delta n}_{k,\ell}(x',z',\cdot)}\big)
    e^{-2\pi i (k\xi + \ell\zeta)} \\
    &\qquad\qquad+\mathbb{E}\big( \widehat{\delta n}_{k,\ell} (x,z,\cdot) 
        \widehat{\delta n}_{0,0}(x',z',\cdot)\big)
    e^{2\pi i (k\xi + \ell\zeta)} \\
    &\qquad\qquad+\mathbb{E}\big( \overline{\widehat{\delta n}_{k,\ell} (x,z,\cdot)} 
        \widehat{\delta n}_{0,0}(x',z',\cdot) \big)
    e^{-2\pi i (k\xi + \ell\zeta)} \Big] \\
    &+\sum_{(k,\ell)\in\mathcal{M}} \sum_{(k',\ell') \in \mathcal{M}} 
    \Big[\mathbb{E}\big( 
        \widehat{\delta n}_{k,\ell}(x,z,\cdot)\, \widehat{\delta n}_{k',\ell'}(x',z',\cdot) \big)
    e^{2\pi i (k\xi + \ell\zeta + k'\xi' + \ell'\zeta')} \\
    &\qquad\qquad\qquad\qquad +\mathbb{E}\big(\widehat{\delta n}_{k,\ell}(x,z,\cdot)\, 
        \overline{\widehat{\delta n}_{k',\ell'}(x',z',\cdot)} \big)
    e^{2\pi i (k\xi + \ell\zeta - k'\xi' - \ell'\zeta')} \\
    &\qquad\qquad\qquad\qquad +\mathbb{E}\big( \overline{\widehat{\delta n}_{k,\ell}(x,z,\cdot)}\, 
        \widehat{\delta n}_{k',\ell'}(x',z',\cdot) \big)
    e^{-2\pi i (k\xi + \ell\zeta - k'\xi' - \ell'\zeta')} \\
    &\qquad\qquad\qquad\qquad +\mathbb{E}\big( \overline{\widehat{\delta n}_{k,\ell}(x,z,\cdot)}\, 
        \overline{\widehat{\delta n}_{k',\ell'}(x',z',\cdot)} \big)
    e^{-2\pi i (k\xi + \ell\zeta + k'\xi' + \ell'\zeta')} \Big].
\end{align*}

The correlation function $C(x,z,x',z')$ is also manifestly real-valued. Again the actual values of the expectation-value operators involved in the foregoing expression depend on the choice of the probability space $(A, \mathcal{F}, \mu)$. A convenient way to generate the statistics of the field makes use of random phases.

## Method of random phases

Probably the simplest way to generate the harmonics, the random phase method consists in the Ansatz

\begin{align*}
    &\widehat{\delta n}_{0,0} (x,z,\alpha) = \Phi_{0,0} (x,z) \cos(2\pi \alpha_{0,0}), & \\
    &\widehat{\delta n}_{k,\ell} (x,z,\alpha) = \Phi_{k,\ell} (x,z) e^{2\pi i \alpha_{k,\ell}},& 
    \quad \text{for $(k,\ell) \in \mathcal{M}$}.
\end{align*}

where $\Phi_{k,\ell}(x,z)$ are deterministic functions over $\Omega$, and $\alpha_{k,\ell}$ are independent random numbers uniformly distributed in the interval $[0,1]$. Since a random phase shift is added, we can take the functions $\Phi_{k,\ell}(x,z)$ to be real-valued, and we only need to define the independent harmonics $(k,\ell) \in \{(0,0)\} \cup \mathcal{M}$. Therefore, the configuration space for $\alpha = (\alpha_{k,\ell})_{(k,\ell) \in \{0,0)\} \cup \mathcal{M}}$ is the unit hypercube $A = [0,1]^N$, where $N=1+|\mathcal{M}|$, $|\mathcal{M}|$ being the number of elements of $\mathcal{M}$; then, $\mathcal{F}$ is the usual Borel $\sigma$-algebra of $A$, and $\mu$ is the uniform probability measure $d\mu(\alpha) = \prod_{(k,\ell) \in \mathcal{M}} d\alpha_{k,\ell}$. The $(0,0)$-harmonic is real-valued by definition.

The zero-average condition $\mathbb{E}(\delta n) = 0$ is automatically satisfied as

\begin{align*}
    &\mathbb{E}\big(\cos(2\pi \alpha_{0,0}) \big) = 
    \int_A \cos(2\pi \alpha_{0,0}) d\mu(\alpha) = 0, \\
    &\mathbb{E}\big(\exp(2\pi i \alpha_{k,\ell}) \big) = 
    \int_A \exp(2\pi i \alpha_{k,\ell}) d\mu(\alpha) = 0.
\end{align*}

It remains to compute the two-point correlation function in order to find the deterministic amplitudes $\Phi_{k,\ell}(x,z)$.

### The two-point correlation function

In order to compute the two-point correlation function, we need the following integrals:

\begin{align*}
    \int_A \cos^2(2\pi \alpha_{0,0}) d\mu(\alpha) 
    &= \int_0^1 \cos^2(2\pi \alpha_{0,0})d\alpha_{0,0} = 1/2, \\
    \int_A \cos(2\pi \alpha_{0,0}) e^{\pm 2\pi i \alpha_{k,l}} d\mu(\alpha) 
    &= \int_0^1 \cos (2\pi\alpha_{0,0}) d\alpha_{0,0} 
    \int_0^1 e^{\pm 2\pi i \alpha_{k,l}} d\alpha_{k,\ell} = 0, \\
    \int_A e^{2\pi i (\alpha_{k,\ell} + \alpha_{k',\ell'})} d\mu(\alpha)
    &= \int_A e^{-2\pi i (\alpha_{k,\ell} + \alpha_{k',\ell'})} d\mu(\alpha)
    = 0, \\
    \int_A e^{2\pi i (\alpha_{k,\ell} - \alpha_{k',\ell'})} d\mu(\alpha)
    &= \int_A e^{-2\pi i (\alpha_{k,\ell} - \alpha_{k',\ell'})} d\mu(\alpha)
    = \delta_{kk'} \delta_{\ell\ell'}.
\end{align*}
    
Correspondingly, the correlation function reads

\begin{align*}
    C(x,z,x',z') = \frac{1}{2} \Phi_{0,0}(x,z) \Phi_{0,0}(x',z')
    + \sum_{(k,\ell) \in \mathcal{M}}  \Big[
    &\Phi_{k,l}(x,z) \Phi_{k,\ell}(x',z')
    e^{2\pi i \big[ k(\xi - \xi') + \ell (\zeta - \zeta') \big]} \\
    &\quad +\Phi_{k,l}(x,z) \Phi_{k,\ell}(x',z')
    e^{-2\pi i \big[k (\xi - \xi') + \ell (\zeta - \zeta')\big]} \Big].
\end{align*}

Again we observe that the correlation function is manifestly real-valued and depends on the difference $(\xi - \xi'), (\zeta - \zeta')$ rather than on $(\xi, \zeta, \xi', \zeta')$ separately. This simplification is the main motivation for working with the Fourier basis (although, in principle other basis could be used). If we define the "negative harmonics" by

\begin{equation*}
    \Phi_{k,\ell} (x,z) = \Phi_{-k,-\ell}(x,z), \quad
    \text{for $(-k, -\ell) \in \mathcal{M}$},
\end{equation*}

where $\Phi_{k,\ell}(x,z)$ are assumed to be real valued are given for 
$(k,\ell) \in \mathcal{M}$, we can write the correlation function in the compact form

\begin{equation*}
    C(x,z,x',z') = \frac{1}{2} \Phi_{0,0}(x,z) \Phi_{0,0}(x',z')
    +\sum_{0 < |k| \leq M_x} \sum_{0 < |\ell| \leq M_z}
    \Phi_{k,\ell} (x,z) \Phi_{k,\ell} (x',z') 
    e^{2\pi i [k(\xi - \xi') + \ell (\zeta - \zeta')]}.
\end{equation*}

The amplitudes $\Phi_{k,\ell}(x,z)$ must be selected so that the corresponding Wigner transform of the correlation matches or at least approximates the desired model.

### The Wigner function

Explicitly, we have

\begin{align*}
    \Gamma(k_0,x,z,N_x,N_z) &= \frac{1}{2}
    \int e^{- i k_0 (N_x s_x + N_z s_z)}
    \Phi_{0,0} (x + \tfrac{1}{2}s_x, z + \tfrac{1}{2} s_z) 
    \Phi_{0,0} (x - \tfrac{1}{2}s_x,z - \tfrac{1}{2}s_z) ds_x ds_z \\
    &\quad + \sum_{0<|k|\leq M_x} \sum_{0<|\ell|\leq M_z} 
    \int e^{- i k_0 (N_x s_x + N_z s_z)}
    \Phi_{k,\ell} (x + \tfrac{1}{2}s_x, z + \tfrac{1}{2} s_z) 
    \Phi_{k,\ell} (x - \tfrac{1}{2}s_x,z - \tfrac{1}{2}s_z) \\
    &\qquad\qquad\qquad\qquad\qquad \times e^{2\pi i [k s_x / a_x + \ell s_z/a_z]} ds_x ds_z.
\end{align*}

As usual we are interested in the principal part of the Wigner function, i.e., in the leading order in the formal high-frequency limit, namely, $k_0 \to + \infty$ (more precisely, the limit parameter is $k_0 L \to +\infty$ where $L$ is the length scale of the problem). Taylor's formula with integral remainder gives

\begin{equation*}
    \Phi_{k,\ell}(x + \tfrac{1}{2}s_x, z + \tfrac{1}{2}s_z) = 
    \Phi_{k,\ell}(x,z) + s_x R_x(x,z,s_x,s_z) + s_z R_z(x,z,s_x,s_z),
\end{equation*}

where $R_x$ and $R_z$ are the remainders. By using the identity

\begin{equation*}
    s_x e^{-ik_0 N_x s_x} = \frac{i}{k_0} \frac{\partial}{\partial N_x} e^{-ik_0 N_x s_x},
\end{equation*}

and the analogous for $s_z$ together with formal integration by parts, we see that remainders corresponds to integrals of the same type but multiplied by $1/k_0$. The Wigner function therefore amounts to 

\begin{align*}
    \Gamma(k_0,x,z,N_x,N_z) &= \frac{1}{2} [\Phi_{0,0}(x,z)]^2 
    \int e^{- i k_0 (N_x s_x + N_z s_z)} ds \\ 
    & \quad +\sum_{0<|k|\leq M_x} \sum_{0<|\ell|\leq M_z}
    [\Phi_{k,\ell} (x, z)]^2
    \int e^{- i k_0 (N_x s_x + N_z s_z)}  
    e^{2\pi i [k s_x / a_x + \ell s_z/a_z]}  ds_x ds_z \\
    & \quad + k_0^{-1} \Gamma_R (k_0,x,z,N_x,N_z),
\end{align*}

where $\Gamma_R$ is the contribution due to the remainders. The integral is now readily evaluated by

\begin{equation*}
    \int e^{- i k_0 (N_x - \frac{2\pi k}{k_0 a_x}) s_x} ds_x =
    \frac{2\pi}{k_0} \delta(N_x - \tfrac{2\pi k}{k_0 a_x} ),
\end{equation*}

and analogously for the integral in $ds_z$. At last, the Wigner function reads,

\begin{align*}
    \Gamma(k_0,x,z,N_x,N_z) = \frac{4 \pi^2}{k_0^2} &\Big[ 
    \frac{1}{2} [\Phi_{0,0} (x,z) ]^2 \delta(N_x)\delta(N_z) \\
    & \qquad +\sum_{0<|k|\leq M_x} \sum_{0<|\ell|\leq M_z}
    [\Phi_{k,\ell} (x, z)]^2 \delta(N_x - \tfrac{2\pi k}{k_0 a_x} )
    \delta(N_z - \tfrac{2\pi \ell}{k_0 a_z} )\Big] \\
    &+ k_0^{-1} \Gamma_R (k_0,x,z,N_x,N_z).
\end{align*}

This is the final form of the Wigner function for the considered model of fluctuations. One can observe that the leading order term amounts to a Dirac's comb, which is a measure. This is expected due to the fact that the considered fluctuations are periodic with spatial periods $a_x$ and $a_z$ in the two directions. 

We recall that the derivation of the wave kinetic equation assumes high regularity for $\Gamma$ as a function of the refractive index. Therefore, The Dirac's comb cannot be used to build a scattering operators in the wave kinetic equation. It is however possible to construct a regular Wigner function which approaches the Dirac's comb in the sense of distributions in the high-frequency limit.

### Approximation with a regular measure and definition of $\Phi_{k,\ell}(x,z)$

Let $\Phi(k_0,x,z,N_x,N_z)$ be a smooth real-valued function with compact support in the box

\begin{equation*}
    \hat{\Omega} = [-N^*_x, N^*_x] \times [-N^*_z, N^*_z],
\end{equation*}

satisfying the duality relation

\begin{equation*}
    \frac{k_0 a_x}{2\pi} N^*_x \in \mathbb{N}, \quad
    \frac{k_0 a_z}{2\pi} N^*_z \in \mathbb{N}.
\end{equation*}

Here, the box is fixed, i.e., both $N_x^*$ and $N_z^*$ do not depend on the frequency. In practice one picks a rapidly decreasing function $\Phi(k_0,x,z,N_x,N_z)$ and select $N_x^*, N_z^*$ sufficiently large to include the approximate support (cutting out small values) of $\Phi(k_0,x,z,.)$ for $(x,z) \in \Omega$. Then, we define the maximum number of harmonics by setting $M_x$ and $M_z$ to the foregoing integer numbers, namely,

\begin{equation*}
    M_x = \frac{k_0 a_x}{2\pi} N^*_x \in \mathbb{N}, \quad
    M_z = \frac{k_0 a_z}{2\pi} N^*_z \in \mathbb{N}.
\end{equation*}

As a result, the number of harmonics grows linearly with the frequency, which enters through the parameter $k_0$, while the size of the box $\hat{\Omega}$ is independent of $k_0$. 

Let us consider the integral

\begin{equation*}
    \langle \chi, [\Phi(k_0,x,z,\cdot)]^2 \rangle =
    \int \chi(N_x,N_z) [\Phi(x,z,N_x,N_z)]^2 dN_x dN_z,
\end{equation*}

where $\chi(N_x, N_z)$ is a test function of class $C^\infty$, i.e., we consider 
$|\Phi(k_0,x,z,\cdot)|^2$ as a compactly supported distribution.

This integral can be approximated by Riemann quadrature formula on the uniform mesh of points

\begin{equation*}
    N_{x,k} = \frac{2\pi k}{k_0  a_x}, \quad
    N_{z,\ell} = \frac{2\pi \ell}{k_0  a_z},
\end{equation*}

for $k \in \{-M_x, \ldots, M_x\}$ and $\ell \in \{-M_z, \ldots, M_z\}$. The step sizes are

\begin{equation*}
    \Delta N_x = \frac{2\pi}{k_0 a_x}, \quad
    \Delta N_z = \frac{2\pi}{k_0 a_z}.
\end{equation*}

The integral over the total domain $\hat{\Omega}$ is written as the sum of integrals over the grid cells 

\begin{equation*}
    \hat{\Omega}_{k,\ell} = 
    \Big[N_{x,k}-\frac{\Delta N_x}{2}, N_{x,k} + \frac{\Delta N_x}{2} \Big] 
    \times \Big[N_{z,\ell} - \frac{\Delta N_z}{2}, N_{z,\ell} + \frac{\Delta N_z}{2}\Big],
\end{equation*}

centered around the $(k,\ell)$ point, and we make use of the Taylor formula with integral remainder for the product $\chi(\cdot) [\Phi(k_0,x,z,\cdot)]^2$, namely,

\begin{align*}
    \chi(N_x,N_z)[\Phi(k_0,x,z,N_x,N_z)]^2 &= 
        \chi(N_{x,k},N_{z,\ell}) [\Phi(k_0,x,z,N_{x,k},N_{z,\ell})]^2 \\ 
     &\qquad\qquad+ (N_x - N_{x,k}) R_{x,k\ell} (k_0,x,z,N_x,N_z) \\
     &\qquad\qquad+ (N_z - N_{z,\ell}) R_{z,k\ell} (k_0,x,z,N_x,N_z),
\end{align*}

with remainders

\begin{align*}
    R_{x,k\ell} (k_0,x,z,N_x,N_z) &= \int_0^1 \partial_{N_x}(\chi(N_x,N_z) [\Phi(k_0,x,z,N_x,N_z)]^2)
        \bigg|_{\substack{N_x=N_{x,k} + t(N_x-N_{x,k})\\ 
                          N_z = N_{z,\ell} + t(N_z - N_{z,\ell})}} dt, \\
    R_{z,k\ell} (k_0,x,z,N_x,N_z) &= \int_0^1 \partial_{N_z}(\chi(N_x,N_z)[\Phi(x,z,N_x,N_z)]^2)
        \bigg|_{\substack{N_x=N_{x,k} + t(N_x-N_{x,k})\\ 
                          N_z=N_{z,\ell} + t(N_z - N_{z,\ell})}} dt.
\end{align*}

We obtain the exact identity

\begin{align*}
    \langle \chi,[\Phi(k_0,x,z,\cdot)]^2 \rangle = \sum_{k,\ell}\bigg[ \chi(N_{x,k}, N_{z,\ell})
    &[\Phi(k_0,x,z,N_{x,k},N_{z,\ell})]^2 \Delta N_x \Delta N_z \\
    &+ I_{x,k\ell}(\chi) \Delta N_x^2 \Delta N_z + 
    I_{z,k\ell} (\chi) \Delta N_x \Delta N_z^2\bigg],
\end{align*}

where $(k,\ell)$ run over the considered mesh so that the sum is finite, and

\begin{align*}
    I_{x,k\ell}(\chi) &= \frac{1}{\Delta N_x \Delta N_z} \int_{\hat{\Omega}_{k\ell}}
    \frac{N_x - N_{x,k}}{\Delta N_x} R_{x,k\ell}(k_0,x,z,N_x,N_y)dN_xdN_z, \\
    I_{z,k\ell}(\chi) &= \frac{1}{\Delta N_x \Delta N_z} \int_{\hat{\Omega}_{k\ell}}
    \frac{N_z - N_{z,\ell}}{\Delta N_z} R_{x,k\ell}(k_0,x,z,N_x,N_y)dN_xdN_z,
\end{align*}

are two linear continuous functionals depending of the mesh sizes $\Delta N_x, \Delta N_z$. At last, we have the exact expression

\begin{equation*}
    \langle \chi, [\Phi(k_0,x,z,\cdot)]^2 \rangle = \frac{4\pi^2}{k_0^2} \sum_{k,\ell}
    \bigg[ \chi(N_{x,k}, N_{z,\ell}) \frac{[\Phi(k_0,x,z,N_{x,k},N_{z,\ell})]^2}{a_x a_z} + 
    \frac{1}{k_0} \Big[\frac{2\pi}{a_x} I_{x,k\ell}(\chi) + 
    \frac{2\pi}{a_z} I_{z,k\ell} (\chi) \Big] \bigg].
\end{equation*}

With those definitions both $I_{x,k\ell}$ and $I_{z,k\ell}$ are bounded in $k_0$, so that for large $k_0$ their contribution is small with respect to the leading term. On the other hand, upon testing $\Gamma$ against $\chi$, we have

\begin{align*}
    \langle \chi, \Gamma(k_0,x,z,\cdot)\rangle &= \frac{4 \pi^2}{k_0^2} \Big[
    \frac{1}{2} \chi(0,0) [\Phi_{0,0}(x,z)]^2 +
    \sum_{0<|k|\leq M_x} \sum_{0<|\ell|\leq M_z} 
    \chi (N_{x,k}, N_{z,\ell}) [\Phi_{k,\ell} (x, z)]^2 \Big] \\ 
    &\quad + \frac{1}{k_0} \langle \chi, \Gamma_R(k_0,x,z,\cdot)\rangle,
\end{align*}

since the mesh points $(N_{x,k}, N_{z,\ell})$ coincide with the support of the principal part in $\Gamma(k,0,x,z,\cdot)$.

We want to choose the smooth function $\Phi(k_0,x,z,N_x,N_z)$ so that the two foregoing expressions agree to lowest order in $1/k_0$, that is,

\begin{align*}
    \Phi_{0,0} (x,z) &= \sqrt{\frac{2}{a_x a_z}} \Phi(k_0,x,z,0,0), \\
    \Phi_{k,\ell} (x,z) &= \frac{1}{\sqrt{a_x a_z}} \Phi(k_0,x,z,N_{x,k},N_{z,\ell}).
\end{align*}

We notice that $\Phi$ must be such that _(a)_ the reality condition $\Phi_{k,\ell} = \Phi_{-k,-\ell}$ is satisfied and _(b)_ $\Phi$ evaluated on grid points should not depend on $k_0$ since the amplitudes of the density fluctuations have no relation with the considered frequency; this means that $\Phi$ must be a function of the wave numbers $k_0 N_x$ and $k_0 N_z$. 

Provided that we can find such a function $\Phi$, we have the exact identity,

\begin{equation*}
    \Gamma(k_0,x,z,N_x,N_z) = [\Phi(k_0,x,z,N_x,N_z)]^2 + \frac{1}{k_0} \Gamma_1,
\end{equation*}

in the weak sense, testing with a smooth function $\chi$. Here, the first order term is defined as the functional

\begin{equation*}
    \langle \chi, \Gamma_1 \rangle = \langle \chi, \Gamma_R(k_0,x,z,\cdot)\rangle
    - \frac{4\pi^2}{k_0^2} 
    \Big[\frac{2\pi}{a_x} I_{x,k\ell}(\chi) + \frac{2\pi}{a_z} I_{z,k\ell} (\chi) \Big].
\end{equation*}

This calculation shows that, in the high frequency regime, the exact Wigner function $\Gamma$ for the considered fluctuation model can be written as a smooth compactly supported function plus a lowest order remainder of $O(1/k_0)$, this identity being valid in the weak sense only (with smooth test functions, i.e., in the space of compactly supported distributions). 

This result allows us to represent the density fluctuations with a Dirac's comb spectral distribution in the framework of the wave kinetic equation and of the WKBeam code in particular.

A standard assumption for the lowest order term in $\Gamma$ is the Gaussian

\begin{equation*}
    \Gamma_{\text{WKBeam}} (k_0,x,z,N_x,N_z) = 
    2\pi \big[n_0(x,z) \langle\frac{\delta n}{n_0}\rangle F(x,z)\big]^2 L_\perp^2 
    e^{-\frac{L_\perp^2}{2} (k_0^2 N_x^2 + k_0^2 N_z^2)},
\end{equation*}

which is a function of the wave numbers $k_0 N_x, k_0 N_z$ only, is real-valued, and symmetric in $N_x , N_z$, and thus it satisfies the requirements _(a)_ and _(b)_ above. Here, $n_0(x,z)$ is the background field, and $F(x,z)$ is the envelope of the fluctuations. This is assumed to take values in [0,1] and it allows us to describe the spatial localization of turbulent fluctuations. The strength of the fluctuations is determined by the root mean square $\langle \delta n/n_0 \rangle$ relative to the background. The scale-length $L_\perp$ is the perpendicular correlation length of the turbulence (and it can in general depend on the point $(x,z)$, but this level of generality is not exploited in the following).

If we set $[\Phi(k_0,x,z,N_x,N_z)]^2 = \Gamma_{\text{WKBeam}}(k_0,x,z,N_x,N_z)$, the amplitudes of the Fourier series of the fluctuation model amount to

\begin{align*}
    \Phi_{0,0}(x,z) &= \sqrt{\frac{4\pi}{a_x a_z}} n_0(x,z) \langle\frac{\delta n}{n_0}\rangle
    F(x,z) L_\perp, \\
    \Phi_{k,\ell} (x,z) &= \sqrt{\frac{2\pi}{a_x a_z}} n_0(x,z) 
    \langle\frac{\delta n}{n_0}\rangle F(x,z) 
    L_\perp e^{-\frac{L_\perp^2}{a_x^2} \pi^2 k^2 - \frac{L_\perp^2}{a_z^2} \pi^2 \ell^2}.
\end{align*}

This result, together with the random phases, fully characterizes the turbulence model. In the following, the foregoing procedure is implemented in a Python class for the generator of samples of the random field $\delta n(x,z)$ with the Gaussian amplitudes given above.

### Summary on the results of the random phase method

The fluctuating field is written in the form

\begin{equation*}
    \delta n(x,z) = n_0(x,z) \langle \frac{\delta n}{n_0} \rangle F(x,z) \Big[
    \tilde{\Phi}_{0,0} \cos( 2 \pi \alpha_{0,0} ) +
    \sum_{(k,\ell) \in \mathcal{M}}
    \tilde{\Phi}_{k,\ell} \; 2\cos\big(2\pi(k\xi + \ell \zeta + \alpha_{k,\ell})\big)
    \Big],
\end{equation*}

where the independent harmonics have been singled out (thus avoiding unnecessarily long loops) and the amplitudes have been re-written in terms of

\begin{align*}
    &\tilde{\Phi}_{0,0} = \sqrt{\frac{4}{\pi} \sigma_x \sigma_z}, \\
    &\tilde{\Phi}_{k,\ell} = \sqrt{\frac{2}{\pi} \sigma_x \sigma_z} 
    \; e^{-\sigma_x^2 k^2 - \sigma_z^2 \ell^2},
\end{align*}
    
where 

\begin{equation*}
    \sigma_x = \pi L_\perp / a_x, \quad
    \sigma_z = \pi L_\perp / a_z,
\end{equation*}

are dimension-less parameters accounting for the ratio of the correlation length to the box sizes.
One can note that each realization of this random field is smooth by definition.

By construction the mean field is zero, namely,

\begin{equation*}
    \mathbb{E}(\delta n) = 0.
\end{equation*}

Since the field is smooth, we can compute its square and thus the variance,

\begin{equation*}
    \mathbb{V}(\delta n) = \mathbb{E}(\delta n^2) - \mathbb{E}(\delta n)^2 =
    \mathbb{E}(\delta n^2),
\end{equation*}

or equivalently

\begin{equation*}
    \mathbb{V}\big(\delta n(x,z)\big) = C(x,z,x,z),
\end{equation*}

where $C(x,z,x',z')$ is the correlation function: The variance is then the restriction of the correlation to the diagonal $x=x'$ $z=z'$. Then, one readily finds

\begin{align*}
    \mathbb{V}\big(\delta n(x,z) \big) &= \frac{1}{2} [\Phi_{0,0}(x,z)]^2
    +\sum_{0 < |k| \leq M_x} \sum_{0 < |\ell| \leq M_z} [\Phi_{k,\ell} (x,z)]^2, \\
    &=[n_0(x,z) \langle \frac{\delta n}{n_0} \rangle F(x,z)]^2 \Big[
    \frac{1}{2} [\tilde{\Phi}_{0,0}]^2
    +\sum_{0 < |k| \leq M_x} \sum_{0 < |\ell| \leq M_z}
    [\tilde{\Phi}_{k,\ell} (x,z)]^2 \Big] \\
    &= [n_0(x,z) \langle \frac{\delta n}{n_0} \rangle F(x,z)]^2
    \frac{2}{\pi} \sigma_x \sigma_z \sum_{k,\ell} e^{-2\sigma_x^2 k^2 - 2\sigma_z^2 \ell^2}.
\end{align*}

The last sum can be approximated by an integral with the result that

\begin{equation*}
    \frac{2}{\pi} \sigma_x \sigma_z \sum_{k,\ell} e^{-2\sigma_x^2 k^2 - 2\sigma_z^2 \ell^2} 
    \approx
    \frac{2}{\pi} \int e^{-2 (u^2 + v^2)} dudv = \frac{1}{\pi} \int e^{-(u^2 + v^2)} dudv = 1.
\end{equation*}

We can therefore write an estimator for the variance, namely,

\begin{equation*}
    \mathbb{V}\big( \delta n (x,z) \big) \approx
    [n_0(x,z) \langle \frac{\delta n}{n_0} \rangle F(x,z)]^2.
\end{equation*}

Usually, the envelope $F$ attains values in $[0,1]$ so that we can interpret the parameter 
$\langle \delta n / n_0 \rangle$ as the maximum root mean square of fluctuations.

The maximum value of the fluctuations relative to $n_0$ can be significantly larger than the root mean square. In fact, with $F(x,z) \in [0,1]$ we have the bound,

\begin{align*}
    \Big|\frac{\delta n(x,z)}{n_0(x,z)} \Big| \leq
    \langle \frac{\delta n}{n_0} \rangle \Big[
    \big|\tilde{\Phi}_{0,0}\big| +
    2 \sum_{(k,\ell) \in \mathcal{M}} \big|\tilde{\Phi}_{k,\ell} \big| \Big]
    &= \langle \frac{\delta n}{n_0} \rangle \Big[ \sqrt{\frac{4}{\pi} \sigma_x \sigma_z} 
    + \sqrt{\frac{2}{\pi} \sigma_x \sigma_z} 
    \sum_{k\not=0}\sum_{\ell\not=0} e^{-\sigma_x^2 k^2 - \sigma_z^2 \ell^2} \Big]\\
    &= \langle \frac{\delta n}{n_0} \rangle 
    \sqrt{\frac{2 \pi}{ \sigma_x \sigma_z} } \Big[
    \frac{1}{\pi} \sigma_x \sigma_z \sum_{k,\ell} e^{-\sigma_x^2 k^2 - \sigma_z^2 \ell^2} 
    + \frac{\sqrt{2} - 1}{\pi} \sigma_x\sigma_z \Big].     
\end{align*}

The remaining sum is deterministic and can be evaluated once for all samples, thus providing a sharp bound for the maximum oscillation. Alternatively, we can estimate the sum by an integral as above, 

\begin{equation*}
    \frac{1}{\pi} \sigma_x \sigma_z \sum_{k,\ell} e^{-\sigma_x^2 k^2 - \sigma_z^2 \ell^2}
    \approx \frac{1}{\pi} \int e^{-(u^2 + v^2)} du dv = 1,
\end{equation*}

thus obtaining the approximate bound

\begin{equation*}
    \Big|\frac{\delta n(x,z)}{n_0(x,z)} \Big| \leq \approx
    \langle \frac{\delta n}{n_0} \rangle \sqrt{\frac{2 \pi}{\sigma_x \sigma_z}},
\end{equation*}

where "$\leq \approx$" means that the bound is given within the error associated to the approximation of the sum with the corresponding integral. The last term in the square bracket has been dropped as it is of the same order as the error on the sum. Since $\sigma_x$ and $\sigma_z$ are typically small as they amount to essentially the ratios of the correlation length to the box sizes, fluctuations can be significantly larger than $\langle \delta n / n_0 \rangle$. 

This also implies that, unless $\langle \delta n / n_0 \rangle$ is sufficiently small, there is a significant probability of negative fluctuations of the total field $n_0 + \delta n$, even when $n_0$ is strictly positive. When the total field $n_0 + \delta n$ represents a density, this is clearly unphysical and should be used with care.

### Testing the sample generator

```python
# Parameters
x1 = -1.
x2 = +1.
z1 = -0.5
z2 = +0.5
Lperp = 0.2
n0 = lambda x,z : ones_like(x)
dn_n0 = 1.
F = lambda x,z : ones_like(x)

# Setting up the test case
parameters = {
    'x1': x1,                
    'x2': x2,                
    'z1': z1,                
    'z2': z2,                
    'Lperp': Lperp,             
    'n0' : n0,
    'dn_n0': dn_n0,
    'F' : F,    
    'tol' : 1.e-10,         
    }

# Construction of the object
deltaN_simple = Fluctuations_RandomPhases(parameters)

# Approximation of fluctuations upper bound
approx_bound = dn_n0 * sqrt((2./pi)*(x2-x1)*(z2-z1)/Lperp**2)

# Some data
Mx = deltaN_simple.Mx
Mz = deltaN_simple.Mz
Nh = deltaN_simple.Nharm
MaxF = deltaN_simple.MaxRelativeFluctuation
print( 'Required max. sample frequency Mx = {}'.format(Mx) )
print( 'Required max. sample frequency Mz = {}'.format(Mz) )
print( 'Number of independent harmonics N = {}'.format(Nh) )
print( 'Max. relative fluctuation = {}'.format(MaxF) )
print( 'Estimated max. relative fluctuation = {}'.format(approx_bound) )
```

```python
# Physics parameters
x1 = 230.         # coordinate of the left boundary of the box in cm
x2 = 250.         # coordinate of the right boundary of the box in cm
z1 = -10.         # lower boundary of the box
z2 = +10.         # upper boundary of the box
Lpedestal = 2.    # scale length (in cm) of the pedestal
Xpedestal = 245.  # coordinate of the center of the pedestal in cm
Npedestal = 1.e13 # density at the pedestal top in cm^-3
dn_n0 = 0.2       # relative level of fluctuations dn/n
Lperp = 0.5       # correlation length in cm
Xfluct = 246.     # position of the turbulence layer
Wfluct = 2.       # width in cm of the turbulence layer

# Background density and fluctuation envelope
n0 = lambda x,z : 0.5 * Npedestal * \
                  (1. - tanh((x-Xpedestal)/Lpedestal))
F = lambda x, z : exp(- (x-Xfluct)**2 / Wfluct**2) 

# Construction of the fluctuation object
parameters ={'x1': x1, 
             'x2': x2, 
             'z1': z1, 
             'z2': z2,
             'Lperp': Lperp,
             'n0' : n0,
             'dn_n0': dn_n0,
             'F' : F,
             'tol' : 1.e-4}
deltaN_AUG = Fluctuations_RandomPhases(parameters)

# Generate the grid
nptx = 200
nptz = 201
X, Z = deltaN_AUG.grid(nptx, nptz)
```

## Full tokamak (ITER - like)

One can generate the density fluctuations in the full tokamak geometry. With this aim, the standard WKBeam equilibrium module is loaded in order to import a given tokamak equilibrium. As an example, the standard ITER case is loaded here.

```python
# Import equilibrium

# ... move to the base of the package to get the modules ...
%cd ../../

# ... needed classes for the plasma
from CommonModules.input_data import InputData 
from CommonModules.PlasmaEquilibrium import TokamakEquilibrium, IntSample

# ... test case to be loaded ...
testcase = 'StandardCases/ITER/ITER0fluct_raytracing.txt'

# ... load equiblibrium ...
idata = InputData(testcase)
background = TokamakEquilibrium(idata)

# ... back to the notebook dir ...
%cd -

```

```python
# Fluctuation model

fl_rho1 = 0.97     # position in rho of the end of the core turbulence
fl_rho2 = 1.0      # position in rho of the start of edge turbulence
coreamp = 0.02     # turbulence level in the core
edgeamp = 0.2      # turbulence level in the edge

                 
def scatteringDeltaneOverne(rho, theta):

    """
    Realistic profile for fluctuations. At the core turbulence level is pretty constant, up to 2 % 
    and at the edge it raises up to 20-80 %. 
    """
    
    profile = coreamp + ((edgeamp - coreamp) / (fl_rho2 - fl_rho1)) * (rho - fl_rho1)
    profile[rho<fl_rho1] = coreamp  # set core value
    profile[rho>=fl_rho2] = edgeamp # set edge value
    
    return profile * (1 + np.cos(theta))
```

```python
# Contruction of the profiles

# ... equilibrium quantities ...
Raxis, Zaxis = background.magn_axis_coord_Rz # coordinates of the magnetic axis
Rmin = background.Rgrid[0,0]
Rmax = background.Rgrid[-1,0]
Zmin = -360.
Zmax = +450.

# ... we need the background profile as a function of rho and theta ...
import scipy.interpolate as spl
psi_profile = background.psi_profile_ne
ne0_profile = background.ne_profile
n0_psi = spl.UnivariateSpline(psi_profile, ne0_profile, s=0.)

# ... define the backgroung density and fluctuation amplitude as functions of R, Z ...
def fluxCoordinates(R, Z):
    RHO = np.sqrt(np.abs(IntSample(R[0,:], Z[:,0], background.PsiInt.eval)))
    THETA = np.arctan2(Z-Zaxis,R-Raxis)
    return RHO, THETA

def n0(R, Z): 
    rho, theta = fluxCoordinates(R, Z)
    return n0_psi(rho.flatten()**2).reshape(rho.shape)

def F(R, Z): 
    rho, theta = fluxCoordinates(R, Z)
    return scatteringDeltaneOverne(rho, theta)

# Construction of the fluctuation object
parameters ={'x1': Rmin, 
             'x2': Rmax, 
             'z1': Zmin, 
             'z2': Zmax,
             'Lperp': 2.0, # isotropic correlation length in cm
             'n0' : n0,
             'dn_n0': 1.,  # set this to one ampitudes are included in F
             'F' : F,
             'tol' : 1.e-4}
deltaN_tok = Fluctuations_RandomPhases(parameters)
```

```python
# Generate the grid (5 points per correlation length)
nptR = 1500
nptZ = 2025
R, Z = deltaN_tok.grid(nptR, nptZ)

# Flux coordinates
RHO, THETA = fluxCoordinates(R, Z) 

# Sample the total field
N = deltaN_tok.full(R, Z)

# Save results (they are too expensive to be regenerated often)
filename = 'ITER_total_density'
np.savez(filename, R=R, Z=Z, density=N)

# plot of the full density (profile plus fluctuations) 
figure(4, figsize=(8,8))
subplot(111, aspect='equal')
pcolormesh(R, Z, N, cmap='hot', vmin=0.)
cDensity = colorbar(orientation='vertical')
cDensity.set_label(r'$n_\mathrm{e}$ [$10^{13} \mathrm{cm}^{-3}$]', fontsize=20)
#contour(R, Z, RHO, [1.0], colors='w', linewidths=3)
xlim(Rmin, Rmax)
ylim(Zmin, Zmax)
xlabel('$R$', fontsize=20)
ylabel('$Z$', fontsize=20)
title('total density', fontsize=20)
grid('on')
savefig('fluctuations_ITERstandard.png', bbox_inches='tight')
```

## Full tokamak (ASDEX Upgrade- like)

The same exercise but for the ASDEX tokamak.

```python
# Import equilibrium

# ... move to the base of the package to get the modules ...
%cd ../../

# ... needed classes for the plasma
from CommonModules.input_data import InputData 
from CommonModules.PlasmaEquilibrium import TokamakEquilibrium, IntSample

# ... test case to be loaded ...
testcase = 'StandardCases/AUG25485/AUG0fluct_raytracing.txt'

# ... load equiblibrium ...
idata = InputData(testcase)
background = TokamakEquilibrium(idata)

# ... back to the notebook dir ...
%cd -

```

```python
# Fluctuation model

fl_rho1 = 0.97     # position in rho of the end of the core turbulence
fl_rho2 = 1.0      # position in rho of the start of edge turbulence
coreamp = 0.02     # turbulence level in the core
edgeamp = 0.2      # turbulence level in the edge

                 
def scatteringDeltaneOverne(rho, theta):

    """
    Realistic profile for fluctuations. At the core turbulence level is pretty constant, up to 2 % 
    and at the edge it raises up to 20-80 %. 
    """
    
    profile = coreamp + ((edgeamp - coreamp) / (fl_rho2 - fl_rho1)) * (rho - fl_rho1)
    profile[rho<fl_rho1] = coreamp  # set core value
    profile[rho>=fl_rho2] = edgeamp # set edge value
    
    return profile * (1 + np.cos(theta))
```

```python
# Contruction of the profiles

# ... equilibrium quantities ...
Raxis, Zaxis = background.magn_axis_coord_Rz # coordinates of the magnetic axis
Rmin = background.Rgrid[0,0]
Rmax = background.Rgrid[-1,0]
Zmin = -100.0
Zmax = +100.0

# ... we need the background profile as a function of rho and theta ...
import scipy.interpolate as spl
psi_profile = background.psi_profile_ne
ne0_profile = background.ne_profile
n0_psi = spl.UnivariateSpline(psi_profile, ne0_profile, s=0.)

# ... define the backgroung density and fluctuation amplitude as functions of R, Z ...
def fluxCoordinates(R, Z):
    RHO = np.sqrt(np.abs(IntSample(R[0,:], Z[:,0], background.PsiInt.eval)))
    THETA = np.arctan2(Z-Zaxis,R-Raxis)
    return RHO, THETA

def n0(R, Z): 
    rho, theta = fluxCoordinates(R, Z)
    return n0_psi(rho.flatten()**2).reshape(rho.shape)

def F(R, Z): 
    rho, theta = fluxCoordinates(R, Z)
    return scatteringDeltaneOverne(rho, theta)

# Construction of the fluctuation object
parameters ={'x1': 200.0, ###Rmin, 
             'x2': 230.0, ###Rmax, 
             'z1': -10.0,  ###Zmin, 
             'z2': 50.0,  ###Zmax,
             'Lperp':1.0, # isotropic correlation length in cm
             'n0' : n0,
             'dn_n0': 1.0,  # set this to one ampitudes are included in F
             'F' : F,
             'tol' : 1.e-4}
deltaN_tok = Fluctuations_RandomPhases(parameters)
```

```python
# Generate the grid (5 points per correlation length)
nptR = 150
nptZ = 202
R, Z = deltaN_tok.grid(nptR, nptZ)

# Flux coordinates
RHO, THETA = fluxCoordinates(R, Z) 

# Sample the total field
N = deltaN_tok.full(R, Z)

# Save results (they are too expensive to be regenerated often)
filename = 'turbAUGdensity_reduced_domain'
np.savez(filename, R=R, Z=Z, density=N)

# plot of the full density (profile plus fluctuations) 
matplotlib.rcParams.update({'font.size': 20})
figure(4, figsize=(8,8))
subplot(111, aspect='equal')
pcolormesh(R, Z, N, cmap='hot', vmin=0.)
cDensity = colorbar(orientation='vertical')
cDensity.set_label(r'$n_\mathrm{e}$ [$10^{13} \mathrm{cm}^{-3}$]', fontsize=30)
#contour(R, Z, RHO, [1.0], colors='w', linewidths=3)
#xlim(Rmin, Rmax)
#ylim(Zmin, Zmax)
xlabel('$R$', fontsize=30)
ylabel('$Z$', fontsize=30)
title('total density', fontsize=30)
grid('on')
savefig('turb_reducedAUGstandard.png', bbox_inches='tight')
```
